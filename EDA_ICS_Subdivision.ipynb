```python
query_avg_division_dataset = """
select 
	record_date,
	avg(idd.extraterrestrial_irradiance) as extraterrestrial_irradiance,
	avg(idd.global_horizontal_irradiance) as global_horizontal_irradiance,
	avg(idd.direct_normal_irradience) as direct_normal_irradience,
	avg(idd.diffuse_horizontal_irradiance) as diffuse_horizontal_irradiance,
	avg(idd.global_horizontal_illumination_klux) as global_horizontal_illumination_klux,
	avg(idd.direct_normal_illumination_klux) as direct_normal_illumination_klux,
	avg(idd.diffuse_horizontal_illumination_klux) as diffuse_horizontal_illumination_klux,
	avg(idd.zenith_illumination) as zenith_illumination,
	avg(idd.sunlight_min) as sunlight_min,
	avg(idd.ceiling_height_meters) as ceiling_height_meters,
	avg(idd.sky_layer_1) as sky_layer_1,
	avg(idd.sky_layer_2) as sky_layer_2,
	avg(idd.sky_layer_3) as sky_layer_3,
	avg(idd.sky_layer_4) as sky_layer_4,
	avg(idd.visibility_km) as visibility_km,
	avg(idd.weather_thunderstorm) as weather_thunderstorm,
	avg(idd.weather_rain) as weather_rain,
	avg(idd.weather_drizzle) as weather_drizzle,
	avg(idd.weather_snow_1) as weather_snow_1,
	avg(idd.weather_snow_2) as weather_snow_2,
	avg(idd.weather_ice) as weather_ice,
	avg(idd.weather_visibility_1) as weather_visibility_1,
	avg(idd.weather_visibility_2) as weather_visibility_2,
	avg(idd.pressure_kpa) as pressure_kpa,
	avg(idd.dry_bulb_temp_c) as dry_bulb_temp_c,
	avg(idd.dew_point_temp_c) as dew_point_temp_c,
	avg(idd.wind_direction_deg) as wind_direction_deg,
	avg(idd.wind_speed_mps) as wind_speed_mps,
	avg(idd.sky_cover) as sky_cover,
	avg(idd.sky_cover_opaque) as sky_cover_opaque,
	avg(idd.snow) as snow 
from 
	ics_division_dataset idd 
group by 
	idd.record_date 
"""
```

```python
dataset = pd.read_sql(query_avg_division_dataset, con=engine, index_col="record_date")
dataset
```

```python


dataset.plot(
    subplots=True,
    title = ["CWEEDS Features Timeseries Plotting"] + ["" for _ in range(30)],
    xlabel = "Timestamp",
    sharex = True,
    figsize=(40,40)
)

```
